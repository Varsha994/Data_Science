```python
### relationship of Age, Fare
plt.scatter(x = df.Age, y =df.Fare)
plt.xlabel('Age')
plt.ylabel('Fare')
# plt.xscale('log')
plt.title('Relation of Age with Fare')
plt.show()
```

```python
sns.scatterplot(df.Age, df.Fare)
plt.title('Relation of Age with Fare')
```

```python
sns.scatterplot(x = 'Age', y = 'Fare', hue = 'Survived', data = df).set_title('Relation of Age with Fare')
```

```python
# plotly, bokeh
#### distribution of Pclass
df.Pclass.unique()
```

```python
y = [df[df.Pclass == 3].shape[0], df[df.Pclass==2].shape[0], df[df.Pclass==1].shape[0]]
plt.bar(x = ['3','2','1'], height= y, color = 'b')
plt.xlabel('Pclass')
plt.ylabel('Count')
plt.show()
```

```python
## subplots
fig, ax = plt.subplots(1,2, figsize = (20,3))
sns.boxplot(df.Age,color = 'r', ax = ax[0]).set_title('Boxplot')
sns.swarmplot(df.Age, ax = ax[1])
ax[1].set_title('Swarmplot')
```

```python
fig, ax = plt.subplots(2,2, figsize = (20,6))
sns.boxplot(df.Age,color = 'r', ax = ax[0,0]).set_title('Boxplot')
sns.swarmplot(df.Age, ax = ax[0,1])
ax[0,1].set_title('Swarmplot')
sns.violinplot(df.Age, ax = ax[1,0])
ax[1,1].scatter(df.Age,df.Fare)
plt.show()
```
